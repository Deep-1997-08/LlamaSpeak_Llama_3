```python
import json
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline
```

```python
config_data=json.load(open('config.json'))
HF_TOKEN=config_data["HF_TOKEN"]
```

```python
model="meta-llama/Meta-Llama-3-8B"
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
tokenizer =AutoTokenizer.from_pretrained(model,token=HF_TOKEN)
tokenizer.pad_token = tokenizer.eos_token
```

```python
model = AutoModelForCausalLM.from_pretrained(model, quantization_config=bnb_config, device_map="auto",token=HF_TOKEN)
```

```python
text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=128)
```

```python
prompt="What is Machine Learning?"
```

```python
response=text_generator(prompt)
```

```python
response
```

```python
def get_response(prompt):
  sequence = text_generator(prompt)
  gen_text = sequence[0]['generated_text']
  return gen_text
```

```python
llama_response=get_response(prompt)
```
